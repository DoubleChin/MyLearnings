# Euler's Method

Hi! In this notebook, I will demonstrate the Euler's method. Enjoy!

Function:
        dy/dx = y, initial condition: y(0) = 1

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
step_size  = 0.001
xs = np.arange(0,10, step_size)
plt.figure(figsize = (5,6))
plt.plot(xs, [math.exp(s) for s in xs])
plt.title("Plot of e^x")
plt.show()
```

Now Let's try to estimate it using dy/dx = y

```python
def euler(x_val, y, step_size, bound):
    slope = y  #This could be varied depending on what function you want to use
    x = x_val
    ys  = np.array([y])
    while x <= bound:
        y_new = y + (step_size * slope)
        ys = np.append(ys,y_new)
        y = y_new
        slope = y #This could be varied depending on what function you want to use
        x+=step_size 
    x = np.arange(x_val, bound, step_size)
    return x, ys
```

```python
step_sizes = [1, 0.1, 0.01, 0.001]
colors = ['red', 'yellow', 'blue', 'orange']
plt.figure(figsize = (10,11))
plt.plot(xs, [math.exp(s) for s in xs], color = 'green', label = 'original', linewidth = 3)
for step_size, color in zip(step_sizes, colors):
    x, y = euler(0,1,step_size,10)   
    plt.plot(x,y[:-2], color = color, label = "Step_size = {}".format(str(step_size)))
plt.legend()
plt.title("Euler's Method", size = 14)
plt.show()
```

### So here, we can see that as we decrease the step size, we also get closer to the estimation of e^x
